```python
import seaborn as sns
```

#### Plotting line chart using Seaborn

```python
# Look how simple seaborn plots line chart

year = [1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010]
unemployment_rate = [9.8, 12, 8, 7.2, 6.9, 7, 6.5, 6.2, 5.5, 6.3]

var = sns.lineplot(x=year, y=unemployment_rate, color="red", marker="o")
var.set(title='unemployment rate vs year', xlabel ="year", ylabel = "unemployment rate")
```

```python
# Look how simple seaborn plots line chart

year = [1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010]
unemployment_rate = [9.8, 12, 8, 7.2, 6.9, 7, 6.5, 6.2, 5.5, 6.3]

sns.lineplot(x=year, y=unemployment_rate, color="red", marker="o").set(title='unemployment rate vs year', xlabel ="year", ylabel = "unemployment rate")
```
