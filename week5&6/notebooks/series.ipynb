## Question: Create series:



```python
# Solution:

name_list = ["Joe", "Nat", "Harry", "Sam", "Monica"]
age_list = [20, 21, 19, 20, 22]
marks_list = [85.10, 77.80, 91.54, 88.78, 60.55]
grade_list = ["A", "B", "A", "A", "B"]
hobby_list = ["Swimming", "Reading", "Music", "Painting", "Dancing"]

index_list = ["S1", "S2", "S3", "S4", "S5"]

name = pd.Series(name_list, index = index_list)
age = pd.Series(age_list, index = index_list)
marks = pd.Series(marks_list, index = index_list)
grade = pd.Series(grade_list, index = index_list)
hobby = pd.Series(hobby_list, index = index_list)
```

### Converting Above Series into a Dataframe

```python
# It supports to combine multiple series.
all_series = [name, age, marks, grade, hobby]

# axis=0 - row
# axis=1 - column
people_data_1 = pd.concat(all_series, axis=1)
people_data_1.head()
```

```python
# Problem: Column names are ambigious (not clear)
```

```python
# Solution: Combining Series by Changing Names
all_dict = {'Name':  name,
             'Age':   age,
             'Marks': marks,
             'Grade': grade,
             'Hobby': hobby}
people_data_2 = pd.concat(all_dict, axis=1)
```

```python
people_data_2.head()
```

```python
# Alternative Solution: 
people_data_3 = pd.DataFrame({'Name':  name,
                   'Age':   age,
                   'Marks': marks,
                   'Grade': grade,
                   'Hobby': hobby})
```
