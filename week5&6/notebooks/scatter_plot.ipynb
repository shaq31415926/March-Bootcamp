## Matplotlib

```python
# Python program to show pyplot module
import matplotlib.pyplot as plt
```

```python
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

```python
import pandas as pd
 
# Reading the CSV file to plot a Histograms chart
iris = pd.read_csv("../data/iris_csv.csv")
iris
```

```python
iris.value_counts()
```

```python
# scatter the sepal_length against the sepal_width
plt.scatter(iris['sepallength'], iris['sepalwidth'])
# set a title and labels
plt.title('Iris Dataset')
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
```

#### Highlighting the points by class

```python
# We can give the graph more meaning by coloring each data point by its class.
# This can be done by creating a dictionary that maps from class to color
# and then use a for-loop to pass the respective color.
```

```python
# create color dictionary
colors = {'Iris-setosa':'r', 'Iris-versicolor':'g', 'Iris-virginica':'b'}
# colors[iris['class'][0]]
# create a figure and axis
fig, ax = plt.subplots()
# plotting each data-point:
# either iterate over sepallength or sepalwidth
# for i in range(iris["sepallegth"].shape[0]):
# for i in range(iris["sepalwidth"].shape[0]):
for i in range(len(iris['sepallength'])):
    ax.scatter(iris['sepallength'][i] , iris['sepalwidth'][i] , color = colors[ iris['class'][i] ])
# set a title and labels
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

```python
# create color dictionary
colors = {'Iris-setosa':'r', 'Iris-versicolor':'g', 'Iris-virginica':'b'}
# colors[iris['class'][0]]
# create a figure and axis
fig, ax = plt.subplots()
# plotting each data-point:
# either iterate over sepallength or sepalwidth
# for i in range(iris["sepallegth"].shape[0]):
for i in range(len(iris['sepallength'])):
    ax.scatter(iris.loc[i, 'sepallength'], iris.loc[i, 'sepalwidth'],color=colors[iris.loc[i, 'class']])
# set a title and labels
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

```python
# create color dictionary
colors = {'Iris-setosa':'r', 'Iris-versicolor':'g', 'Iris-virginica':'b'}
# create a figure and axis
fig, ax = plt.subplots()
# plotting each data-point:
# either iterate over sepallength or sepalwidth
for i in range(len(iris['sepalwidth'])):
    ax.scatter(iris['sepallength'][i], iris['sepalwidth'][i],color=colors[iris['class'][i]])
# set a title and labels
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

```python
# create color dictionary
colors = {'Iris-setosa':'r', 'Iris-versicolor':'g', 'Iris-virginica':'b'}
# colors[iris['class'][0]]
# create a figure and axis
fig, ax = plt.subplots()
# plotting each data-point:
# either iterate over sepallength or sepalwidth
for i in range(iris.shape[0]):
    ax.scatter(iris.loc[i, 'sepallength'], iris.loc[i, 'sepalwidth'],color=colors[iris.loc[i, 'class']])
# set a title and labels
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```
