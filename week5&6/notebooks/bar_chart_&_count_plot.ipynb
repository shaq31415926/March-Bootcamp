# Bar Chart

## Matplotlib

```python
# Python program to show pyplot module
import matplotlib.pyplot as plt
```

```python
import pandas as pd
 
# Reading the CSV file to plot a bar chart
iris = pd.read_csv("../data/iris_csv.csv")
iris
```

```python
# This will plot a simple bar chart
plt.bar(iris['class'], iris['sepallength'])
 
# Title to the plot
plt.title("Iris Dataset")
plt.xlabel("Class")
plt.ylabel("Sepal Length")
 
# Adding the legends
plt.legend(["bar"])
plt.show()
```

```python
# This will plot a simple bar chart
plt.barh(iris['class'], iris['sepallength'])
 
# Title to the plot
plt.title("Iris Dataset")
plt.xlabel("Class")
plt.ylabel("Sepal Length")
 
# Adding the legends
plt.legend(["bar"])
plt.show()
```

### Calculating the frequency of a category

```python
# Reading the CSV file to plot another bar chart
amazon = pd.read_csv("../data/amazon.csv")
amazon
```

```python
# As seen above bar chart do not automatically calculate frequency of a category, 
# therefore, we will use pandas value_counts method to do this.
```

```python
# counting the occurrence of each class of ratings (1-5)
amazon['rating'].value_counts()
```

```python
# count the occurrence of each class of ratings (1-5)
rating_values = amazon['rating'].value_counts() 
# get x and y data 
ratings = rating_values.index 
frequency = rating_values.values 
# create bar chart 
plt.bar(ratings, frequency) 
# set title and labels 
plt.title('Consumer Reviews of Amazon Products') 
plt.xlabel('Rating') 
plt.ylabel('Frequency')
```

## Seaborn

```python
import seaborn as sns
```

```python
sns.barplot(data = iris, x = iris['class'], y = iris['sepallength'])
```

# Count Chart

```python
# Showing frequency of categorical values (1-5) in rating column
sns.countplot(x = 'rating', data = amazon)
```

```python
# Use a different colour palette in count plot
sns.countplot(x = 'rating', data = amazon, palette = "Set2")
```

```python
# Use a different colour palette in count plot
sns.countplot(x = 'rating', data = amazon, color = "yellow")

# or:
#sns.countplot(x = 'rating', data = amazon, color = "#FFFF00")
```

#### Task:

Question: Use "company.csv" data. Read the dataset.

Answer:

```python
# Reading the CSV file to plot a bar chart
company = pd.read_csv("../data/company.csv")
company
```

Question: Create X and y data by selecting appropriate column.

Answer:

```python
x = company.loc[:,"Countries"]
y = company.loc[:,"Companies"]
```

Question: Plot the verticle bar chart using either Matplotlib or Seaborn library.

Answer:

```python
fig, ax = plt.subplots(figsize=(10, 5))
# Plot the data using bar() method
ax.bar(x, y, color='g')
ax.set_title("Innovative companies")
ax.set_xlabel("Countries")
ax.set_ylabel("Number of Companies")
```

Question: Plot the horizontal bar chart.

Answer:

```python
fig, ax = plt.subplots(figsize=(10, 5))
# Plot the data using bar() method
ax.barh(x, y, color='g')
ax.set_title("Innovative companies")
ax.set_xlabel("Number of Companies")
ax.set_ylabel("Countries")
```
