```python
# get correlation matrix
corr = corr_data.corr()
fig, ax = plt.subplots()
# create heatmap
im = ax.imshow(corr.values)

# set labels
ax.set_xticks(np.arange(len(corr.columns)))
ax.set_yticks(np.arange(len(corr.columns)))
ax.set_xticklabels(corr.columns, rotation=45)
ax.set_yticklabels(corr.columns)
```

```python
# get correlation matrix
corr = corr_data.corr()
fig, ax = plt.subplots()
# create heatmap
im = ax.imshow(corr.values)

# set labels
ax.set_xticks(np.arange(len(corr.columns)))
ax.set_yticks(np.arange(len(corr.columns)))
# Rotate the tick labels and set their alignment.
ax.set_xticklabels(corr.columns, rotation=45, ha="right",  rotation_mode="anchor")
ax.set_yticklabels(corr.columns)


# Loop over data dimensions and create text annotations.
for i in range(len(corr.index)): # looping row
    for j in range(len(corr.columns)): # looping columns
        text = ax.text(i, j, np.around(corr.iloc[i, j], decimals=2),
                           ha="center", va="center", color="black")
```

```python
import numpy as np
fig, ax = plt.subplots(figsize=(8,8))
im = ax.imshow(corr_df, interpolation='nearest')
fig.colorbar(im, orientation='vertical', fraction = 0.05)

# Show all ticks and label them with the dataframe column name
ax.set_xticklabels(country_df.columns, rotation=65, fontsize=15)
ax.set_yticklabels(country_df.columns, rotation=0, fontsize=15)

# Loop over data dimensions and create text annotations
for i in range(len(country_df.columns)-1):
    for j in range(len(country_df.columns)-1):
        text = ax.text(j, i, np.around(corr_df.iloc[i, j], decimals=2),
                       ha="center", va="center", color="black")
```
