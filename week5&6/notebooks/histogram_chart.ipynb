# Histogram Chart

## Matplotlib

```python
# Python program to show pyplot module
import matplotlib.pyplot as plt
```

```python
import numpy as np

a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])

plt.hist(a, bins = [0,25,50,75,100])
plt.title("histogram of result")
plt.xticks([0,25,50,75,100])
plt.xlabel('marks')
plt.ylabel('no. of students')
plt.show()
```

```python
import pandas as pd
 
# Reading the CSV file to plot a Histograms chart
iris = pd.read_csv("../data/Iris.csv")
iris
```

```python
plt.hist(iris["sepallength"], bins = 10)
 
# Title to the plot
plt.title("Histogram")
 
# Adding the legends
plt.legend(["Sepal Length (cm)"])
plt.show()
```

```python
# Reading the CSV file to plot another Histograms chart
amazon = pd.read_csv("../data/amazon.csv")
amazon
```

### Plotting Histogram for categorical data

```python
# Histogram for categorical data = bar chart
```

```python
# plot histogram
plt.hist(amazon['rating'], bins = [1,2,3,4,5])
# set title and labels
plt.title('Consumer Reviews of Amazon Products')
plt.xlabel('Ratings')
plt.ylabel('Frequency')
```

## Seaborn

```python
import seaborn as sns
```

```python
# Look bars are total 10 in the histogram chart below
sns.distplot(iris["sepallength"], bins=10, kde=False)
```

```python
# Showing frequency of categorical values (1-5) in rating column
sns.distplot(iris["sepallength"], bins=10, kde=True)
```

```python
# gaussian kernel density estimate (kde) is by default True
sns.distplot(iris["sepallength"], bins=10, kde=True)
```
